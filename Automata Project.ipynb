```python
import pandas as pd

#Taking inputs from user

#number of states
while True:
    try:
        num_of_states = int(input('Please enter number of states: '))
        if num_of_states > 0:
            break
    except:
        pass
    
#states will be represented as q0, q1, ... qn
states = [f'q{q}' for q in range(num_of_states)]

#Initial state
while True:
    try:
        initial_state = input('Enter initial state: ')
        if initial_state in states:
            break
    except:
        pass
    
#Final states
while True:
    try:
        final_states = input('Enter final states (comma separated): ').split(',')
        flag = True
        for final_state in final_states:
            if final_state not in states:
                flag = False
                break
        if flag:
            break
    except:
        pass

#Input symbols
while True:
    try:
        num_of_inputs = int(input('Enter number of inputs: '))
        if num_of_inputs > 0:
            input_symbols = []
            for i in range(num_of_inputs):
                input_symbols.append(input(f'Enter input symbol # {i+1}: '))
            break
    except:
        pass
    
#Transitions
transitions = {}
for i in range(len(states)):
    transitions[states[i]] = {}
    for j in range(len(input_symbols)):
        while True:
            try:
                move_state = input(f'Enter state for {states[i]} at input {input_symbols[j]}: ')
                if move_state in states:
                    transitions[states[i]][input_symbols[j]] = move_state
                    break
            except:
                pass
```

```python
#Hard Coded values for testing

# doesnâ€™t contains 110
states = ['q0', 'q1', 'q2', 'q3']
initial_state = 'q0'
final_states = ['q0', 'q1', 'q2']
input_symbols = ['0', '1']
transitions = {
    'q0': {'0': 'q0', '1': 'q1'},
    'q1': {'0': 'q0', '1': 'q2'},
    'q2': {'0': 'q3', '1': 'q2'},
    'q3': {'0': 'q3', '1': 'q3'}
}
```

```python
#Test string as input
test_string = ''
while True:
    try:
        test_string = input('Enter test string: ')
        test_string = test_string.strip()
        is_valid = False
        if test_string != '':
            is_valid = True
            for i in range(len(test_string)):
                if test_string[i] not in input_symbols:
                    is_valid = False
                    break
        if is_valid:
            break
    except:
        pass

#Check if string is acceptable or not
states_moved = [initial_state]
df_data = []
for i in range(len(test_string)):
    current_state = states_moved[-1]
    next_state = transitions[current_state][test_string[i]]
    states_moved.append(next_state)
    df_data.append([current_state, test_string[i], next_state])

#Check final state
if states_moved[-1] in final_states:
    print('\nString is accepted!')
    print(' --> '.join(states_moved))
else:
    print('\nString is rejected!')
    print(' --> '.join(states_moved))

#Show Pandas DataFrame
df = pd.DataFrame(df_data)
df.columns = ['Current State', 'Input', 'Next State']
display(df)
```
